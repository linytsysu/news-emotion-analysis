```python
model_name = 'bert-base'
model_type = 'preprocess'

content_df = pd.read_csv('../data/Train_DataSet.csv')
label_df = pd.read_csv('../data/Train_DataSet_Label.csv')

df = content_df.merge(label_df, on='id')

df = df.fillna('EMPTY')

# df['titlecontent'] = df['title'] + df['content']
```

```python
df['content'] = [re.sub(r'\s+', ' ', content) for content in df['content'].values]
df['title'] = [title + '\n' for title in df['title'].values]
```

```python
df['titlecontent'] = df['title'] + df['content']
```

```python
for i in range(df.shape[0]):
    text = df['titlecontent'][i]
    text = re.sub(r'\\n+', '。', text)
    text = text.replace('。。', '。')
    text = text.replace('？。', '。')
    text = text.replace('！。', '。')
    text = text.replace(' 。', '。')
    print(i, df['title'][i])
```

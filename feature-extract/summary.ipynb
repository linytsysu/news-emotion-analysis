```python
import numpy as np
import pandas as pd
import re
from itertools import chain
from textrank4zh import TextRank4Keyword, TextRank4Sentence
import zhon.hanzi
from bert_serving.client import BertClient
```

```python
model_name = 'bert_base'
```

```python
bert_client = BertClient()
```

```python
content_df = pd.read_csv('../data/Train_Dataset.csv')
label_df = pd.read_csv('../data/Train_Dataset_Label.csv')
```

```python
train_df = content_df.merge(label_df, on='id')
train_df = train_df.fillna('EMPTY')
train_df.shape
```

```python
sentence_list = []
for i in range(train_df.shape[0]):
    text = re.sub(r'\\n+', '。', str(content_df['content'].values[i]))
    text = text.replace('。。', '。')
    text = text.replace('？。', '。')
    text = text.replace('！。', '。')
    sentence = re.sub(r'[^\u4e00-\u9fa5%s]+'%(zhon.hanzi.punctuation), '', text)

    tr4s = TextRank4Sentence()
    tr4s.analyze(text=sentence, lower=True, source='all_filters')
    result = '。'.join([item.sentence for item in tr4s.get_key_sentences(num=5)])
    sentence_list.append(result)


# sentence = '。'.join([item.sentence for item in tr4s.get_key_sentences(num=5)])
# print(sentence)
# if len(sentence) < 100:
#     print(text)
#     print(''.join(re.findall(zhon.hanzi.sentence, text)))
#     print(sentence)
#     print(content_df['content'].values[65])
```

```python
sentence_list = []
for i in range(train_df.shape[0]):
    text = '。'.join(str(content_df['content'].values[i]).split('\\n'))
    sentence = re.sub(r'[^\u4e00-\u9fa5%s]+'%(zhon.hanzi.punctuation), '', text)
    tr4s = TextRank4Sentence()
    tr4s.analyze(text=sentence, lower=True, source = 'all_filters')
    sentence = '。'.join([item.sentence for item in tr4s.get_key_sentences(num=5)])
    sentence_list.append(sentence)
train_data = bert_client.encode(sentence_list)
print(train_data.shape)
pd.DataFrame(train_data).to_csv('./%s/train_summay_2_word_vector.csv'%(model_name), header=None, index=None)
```

```python
sentence_list = []
for sentence in train_df['content'].values:
    sentence = re.sub(r'[^\u4e00-\u9fa5%s]+'%(zhon.hanzi.punctuation), '', str(sentence))
    sentence_list.append(sentence)
train_data = bert_client.encode(sentence_list)
print(train_data.shape)
pd.DataFrame(train_data).to_csv('./%s/train_content_2_word_vector.csv'%(model_name), header=None, index=None)
```
